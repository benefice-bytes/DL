```
# list out the datasets
tfds.list_builders()
```

```
(train_data, validation_data), test_data = tfds.load(name='imdb_reviews',
         split = (tfds.Split.TRAIN.subsplit([6,4]), tfds.Split.TEST),
         as_supervised = True)
```

```
train_examples_batch, train_labels_batch = next(iter(train_data.batch(10)))

train_examples_batch
```

```
train_labels_batch
```

```
import tensorflow_hub as hub

pre_trained_model = hub.load("https://tfhub.dev/google/tf2-preview/gnews-swivel-20dim/1")

hub_layer = hub.KerasLayer(pre_trained_model, input_shape=[], dtype=tf.string, trainable=True)
```

```
# Here i'm printing 2 reviews 
train_examples_batch[:2] 
```

```
# above 2 moview reviews text converted to 2 vectors
hub_layer(train_examples_batch[:2])
```

```
# create a neural network

model = tf.keras.Sequential()

model.add(hub_layer)  # add input layer
model.add(tf.keras.layers.Dense(16, activation='relu'))
model.add(tf.keras.layers.Dense(1, activation='sigmoid'))

model.summary()
```

```
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```
model.fit(train_data.shuffle(10000).batch(512),
          epochs=20,
          validation_data = validation_data.batch(512),
          verbose = 1)
```

```
model.predict(['This is the worst movie I have ever seen'])
```
